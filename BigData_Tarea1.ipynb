```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SpotifyData").getOrCreate()
```

```python
from pyspark.sql.functions import col
ruta = '/content/drive/MyDrive/SpotifyFeatures.csv'
df = spark.read.csv(ruta, header=True, inferSchema=True)
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
# Canciones del género "dance" lanzadas por Lady Gaga
Dance_LadyGaga = df.filter((col("genre") == "Dance") & (col("artist_name") == "Lady Gaga"))
Dance_LadyGaga.show(5)
```

```python
df.describe(["popularity"]).show()
```

```python
# Conteo por tipo de contenido:
df.groupBy("genre").count().show()
```
